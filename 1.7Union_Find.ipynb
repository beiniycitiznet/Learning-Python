```python
import random

class Graph():
    def __init__(self, num):
        self.nodes = []
        self.edges = []
        for i in range(num):
            self.nodes.append(Node(i))
        for i, node in enumerate(self.nodes):
            for j in range(1):
                next_node = i
                while next_node == i:
                    next_node = random.randint(0, num - 1)
                node.add(self.nodes[next_node])
                self.edges.append((i, next_node))
                self.nodes[next_node].add(node)
    def get_edges(self):
        return self.edges
    def size(self):
        return len(self.nodes)
    
    def print(self):
        for node in self.nodes:
            print(f'{node.val} -> {sorted([x.val for x in node.neighbors])}')
    
            
class Node():
    def __init__(self, val):
        self.val = val
        self.neighbors = set()
    def add(self, node):
        self.neighbors.add(node)
    def add_all(self, nodes):
        for node in nodes:
            self.add(node)
    def node_print(self):
        print(self.val)
```

```python
def union_find(graph):
    edge_list=graph.get_edges()
    map_size=graph.size()
    cluster = []
    cluster_dic={}

    def find(node):
        return cluster[node]
        
    def union(pair):
        fix=find(pair[1])
        correct=find(pair[0])
        for i,value in enumerate(cluster):
            if value==fix:
                cluster[i]=correct
                
    for i in range(map_size):
        cluster.append(i)
        
    for pair in edge_list:
        union(pair)                      

    for i,value in enumerate(cluster):
        if value not in cluster_dic:
            cluster_dic[value]=[i]
        else:
            cluster_dic[value].append(i)
            
    return cluster_dic
```

```python
g = Graph(100)
g.print()
print(union_find(g))
```
